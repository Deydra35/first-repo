Напишите код на Python, реализующий расчет длины вектора, заданного его координатами.

```python
string = input('Введите координаты вектора через пробел:')
vector = string.split()
sum_sqr = 0
for x in vector:
    sum_sqr += int(x)**2
len_vector = math.sqrt(sum_sqr)
print('Длина вектора:', len_vector)
```

1. окружности

```python
r = 1
x1 = []
x2 = []
y1 = []
y2 = []
for i in range(30000):
    x_ = i/30000
    x1.append(x_)
    x2.append(-x_)
    y1.append(math.sqrt(r**2-x_**2))
    y2.append(-math.sqrt(r**2-x_**2))
plt.plot(x1,y1)
plt.plot(x1,y2)
plt.plot(x2,y1)
plt.plot(x2,y2)
plt.xlabel("x")
plt.ylabel("y")
```

4.1 Решите систему уравнений:
y = x2 – 1
exp(x) + x∙(1 – y) = 1

```python
from scipy.optimize import fsolve

x = np.linspace(-2,3,201)
plt.plot(x, x**2 - 1)
plt.plot(x, (np.exp(x)+x-1)/x)
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
```

```python
from scipy.optimize import fsolve

def equations(p):
    x, y = p
    return (y-x**2+1, np.exp(x)+x*(1-y)-1)

x1, y1 = fsolve(equations, (2,4))
print(x1,y1)

x2, y2 = fsolve(equations, (-2,1))
print(x2,y2)
```

4.2 Решите систему уравнений и неравенств:
y = x2 – 1
exp(x) + x∙(1 – y) > 1

```python
from scipy.optimize import fsolve

x = np.linspace(-4,6,201)
plt.plot(x, np.exp(x)+x*(2-x**2)-1)
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
```

```python
from scipy.optimize import fsolve

def f(x):
    return (np.exp(x)+x*(2-x**2)-1)

x0 = fsolve(f,-2)
print(x0)
x1 = fsolve(f,0)
print(x1)
x2 = fsolve(f,2)
print(x2)
x3 = fsolve(f,4)
print(x3)
```
